Part A     load data    preprocess    split 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

#def load_function

def load_data():
    real_sequence=[];fake_sequence=[];total_sequence=[]
    real_txt=open('clean_real.txt',mode='r')
    fake_txt=open('clean_fake.txt',mode='r')
    for w1 in real_txt:
        real_sequence.append(w1)
    for w2 in  fake_txt:
        fake_sequence.append(w2)
    total_sequence=real_sequence+fake_sequence
    vect=CountVectorizer()
    x=vect.fit_transform(total_sequence).toarray()
    y=[]
    for i in range(len(real_sequence)):
        y.append(1)
    for j in range(len(fake_sequence)):
        y.append(0)
    x_tr,x1,y_tr,y1=train_test_split(x,y,train_size=0.7,random_state=42)
    x_val,x_tst,y_val,y_tst=train_test_split(x1,y1,test_size=0.5,random_state=42)
    return x_tr,y_tr,x_val,y_val,x_tst,y_tst

# def  data variables
x_train,y_train,x_validation,y_validation,x_test,y_tst=load_data()
```

part B    KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
def knn_constructor(n):
    global knn
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(x_train,y_train)

train_acc=[]
validation_acc=[]
for i in range(1,21):
    knn_constructor(i)
    train_acc.append(knn.score(x_train,y_train))
    validation_acc.append(knn.score(x_validation,y_validation))
```

k values _ train  set score  plot

```python
import numpy as np
import matplotlib.pyplot as plt
plt.scatter([i for i in range(1,21)],train_acc)
plt.xlabel("k_values")
plt.ylabel("knn train_scores")
plt.show()
```

k_values _ validation set score

```python
import numpy as np
import matplotlib.pyplot as plt
plt.scatter([i for i in range(1,21)],validation_acc)
plt.xlabel("k_values")
plt.ylabel("knn validation_scores")
plt.show()
```
